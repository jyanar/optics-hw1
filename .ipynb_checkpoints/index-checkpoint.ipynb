# Problem #2
Create a program in Excel that will yield the 1931 and 1976 CIE coordinates for a blackbody radiator (the user will enter the blackbody temperature). You can use the formulas given below:

$$
\bar{y} = exp[\frac{-833.202209 + 2.995009\lambda - 0.002691\lambda^2}{1 + 0.011476\lambda + 0.000006\lambda^2}]
$$

$$
\\
$$

$$
\bar{x} = exp[\frac{-9712.539062 + 32.536427\lambda - 0.027238\lambda^2}{1 + 0.270274\lambda - 0.00028\lambda^2}] + exp[\frac{-225.256866 + 1.011834\lambda - 0.001141\lambda^2}{1 - 0.004699\lambda + 0.00001\lambda^2}]
$$

$$
\\
$$

$$
\bar{z} = exp[\frac{0.620450 + 0.000582(\lambda - 450\space nm) - 0.000973(\lambda - 450\space nm)^2}{1 + 0.008606(\lambda - 450\space nm)}]
$$

$$
\\
$$

$$
M_\lambda(\lambda, T) = \frac{2\pi h c^2}{\lambda^5}\frac{1}{e^\frac{hc}{k_B T\lambda - 1}}
$$

$$
h = 6.6261 * 10^-34 \space \frac{J}{Hz}\\
k_B = 1.281*10^-23 \space \frac{J}{K}
$$

```python
global h
global k_B
global c
global lambda_range

h   = 6.261*10**-34
k_B = 1.381*10**-23 
c   = 3.8  *10**8
lambda_range = np.array(range(350, 781))

# The three Color Matching Functions, \bar{y}, \bar{x}, and \bar{z}
def y_bar(lambd):
    return np.exp((-833.202209 + 2.995009*lambd - 0.002691*lambd**2)\
                 /(1 + 0.011476*lambd + 0.000006*lambd**2))

def x_bar(lambd):
    return np.exp((-9712.539062 + 32.536427*lambd - 0.027238*lambd**2)/(1 + 0.270274*lambd - 0.00028*lambd**2))\
         + np.exp((-225.256866 + 1.011834*lambd - 0.001141*lambd**2)/(1 - 0.004699*lambd + 0.00001*lambd**2))

def z_bar(lambd):
    return np.exp((0.620450 + 0.000582*(lambd - 450) - 0.000973*(lambd - 450)**2)
             /(1 + 0.008606*(lambd - 450)))

def M_lambd(lambd, T):
    """ Planck's Law.
    Args:
        lambd [m] - Scalar, represents wavelength.
        T     [K] - Scalar, color temperature. 
    """
    return ( (2 * np.pi * h * c**2)\
            /(((lambd*10**-9)**5) * np.exp(((h*c) / (k_B * T * lambd*10**-9)) - 1)) )

def tristim_X(T):
    return np.sum([M_lambd(L, T) * x_bar(L) for L in lambda_range])

def tristim_Y(T):
    return np.sum([M_lambd(L, T) * y_bar(L) for L in lambda_range])

def tristim_Z(T):
    return np.sum([M_lambd(L, T) * z_bar(L) for L in lambda_range])

def cie1931_x(T):
    return tristim_X(T) / (tristim_X(T) + tristim_Y(T) + tristim_Z(T))

def cie1931_y(T):
    return tristim_Y(T) / (tristim_X(T) + tristim_Y(T) + tristim_Z(T))

def cie1976_u(T):
    return (4*cie1931_x(T)) / (-2*cie1931_x(T) + 12*cie1931_y(T) + 3)

def cie1976_v(T):
    return (9*cie1931_y(T)) / (-2*cie1931_x(T) + 12*cie1931_y(T) + 3)
```

### Color Matching Functions

Now that we've defined $\bar{x}$, $\bar{y}$, and $\bar{z}$, let's plot them:

```python
plt.plot(lambda_range, x_bar(lambda_range), label='$\\bar{x}$', color='red')
plt.plot(lambda_range, y_bar(lambda_range), label='$\\bar{y}$', color='green')
plt.plot(lambda_range, z_bar(lambda_range), label='$\\bar{z}$', color='blue')
plt.xlabel('$\\lambda/nm$'); plt.legend()
```

And Planck's Law, for several different temperatures:

```python
l_range = np.array(np.linspace(0, 4000, 1000))
plt.plot(l_range, M_lambd(l_range, 3000), label='3000 K')
plt.plot(l_range, M_lambd(l_range, 4000), label='4000 K')
plt.plot(l_range, M_lambd(l_range, 5000), label='5000 K')
plt.title("Planck's Law"); plt.legend()
```

##### Let's plot the [Planckian locus](https://en.wikipedia.org/wiki/Planckian_locus) in 1931 and 1976 CIE coordinates. 

```python
x = []; u = []
y = []; v = []
for i in np.linspace(1500, 10000, 100):
    x.append(cie1931_x(i))
    y.append(cie1931_y(i))
    u.append(cie1976_u(i))
    v.append(cie1976_v(i))

plt.plot(x, y, label='CIE 1931')
plt.plot(u, v, label='CIE 1976')
plt.legend()
```

# Problem #4
Use the file `FFS-100-400.txt` to read in the spectral radiant flux of the calibrated light source. Write a program that calculates the luminous flux of the light source.

The luminous flux of the light source is defined as:

$$
X_{\lambda} = \left( 683\space \frac{lm}{W} \right) \sum\limits_{380\space nm}^{780 nm} X_{\lambda}(\lambda) V_{\lambda}(\lambda) \Delta\lambda
$$

```python
df = pd.read_csv('FFS-100-400.csv')
df.head()
```

```python
lum_flux = 683 * np.sum([df.radiant_flux * y_bar(df.n)])
print("The Luminous Flux of the light source, given the spectral radiant flux:\n {:.3f} lm".format(lum_flux))
```

```python
def luminous_flux(lambd, radiant_flux):
    """ Converts radiant flux at a given wavelength to luminous flux.
    Args:
        lambd [nm] - Scalar, wavelength in nanometers.
        radiant_flux [W/nm] - Scalar, radiant flux given in text file. Watts / nanometers.    
    """
    lum_flux = 0
    for L in lambda_range:
        lum_flux += radiant_flux * y_bar(lambd)
    return 683 * lum_flux
```

```python
df['lum_flux'] = luminous_flux(df.n, df.radiant_flux)
df.head()
```

# Problem #5
Again use `FFS-100-400`. Write a program that takes the spectral radiance of the opal glass of the calibrated light source to determine luminous flux of the test light source.

The luminous flux of the test source is defined as:

$$
\Phi_{V,TS} = \sum\limits_{\lambda=580\space nm}^{780\space nm} \Phi_{\lambda,TS} V_{\lambda} \Delta\lambda
$$

where the spectral radiant flux of the test source, $\Phi_{\lambda,TS}$, is defined as

$$
\Phi_{\lambda,TS} = \frac{L_{\lambda,TS}}{L_{\lambda,CS}} \Phi_{\lambda,CS}
$$

and $\Phi_{\lambda,CS}$ are the values we obtain from the text file.

```python
df = pd.read_csv('FFS-100-400.csv')
df.head()
```

```python
def spectral_rad_flux_ts(L_lambd_ts, L_lambd_cs, spectral_rad_flux_cs):
    """ Returns the spectral radiant flux of the test source.
    Args:
        lambd [nm] - Scalar, wavelength in nanometers at given wavelength
        L_lambd_ts - Scalar, luminance of test source at given wavelength
        L_lambd_cs - Scalar, luminance of calibrated source at given wavelength
        spectral_rad_flux_cs - Scalar, rad flux of calibrated source at given wavelength
    """
    return (L_lambd_ts / L_lambd_cs) * spectral_rad_flux_cs

def luminous_flux_ts(df):
    """ Returns luminous flux of a test source given a dataframe (df) with the following columns:
    Columns of df:
        n [nm] - Wavelengths in nanometers
        radiant_flux - Radiant flux of calibrated test source
        ts_lum - Luminance of test source
        cs_lum - Luminance of calibrated source
    """
    return np.sum([spectral_rad_flux_ts(float(df[df.n==L].ts_lum), 
                                        float(df[df.n==L].cs_lum), 
                                        float(df[df.n==L].radiant_flux)
                                       ) * y_bar(L) \
                   for L in range(580, 781)])
```

Since we don't currently have values for $L_{\lambda,CS}$ and $L_{\lambda,TS}$ (luminance), we cannot currently compute the luminous flux of the test source, but we can pretend that

$$
\frac{L_{\lambda,TS}}{L_{\lambda,CS}} = 1,
$$

giving us:

```python
df['ts_lum'] = 1
df['cs_lum'] = 1
df.head()
```
